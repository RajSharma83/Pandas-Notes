```python
import pandas as pd
```

Adding Column

```python
data = {
    "name":['Raj Sharma', 'Surya Pratap', 'Prince Kumar', 'Aniket Yadav', 'Ayush Singh', 'Prince sharma'],
    "age": [19, 20, 21, 32, 24, 22],
   "salray": [20000, 25000, 30000, 40000, 35000, 32000],
    "performance score": ['Good', 'Very Good', 'Excellent', 'Average', 'Good', 'Excellent']
}
df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)

# if we want to add column at the end of dataframe
df["Bonus"] = df["salray"] * 0.10 
print("\nDataFrame after adding Bonus column:")
print(df)

# using insert() - > if we want to add column at specific location
# df.insert(loc, column data , some value)
df.insert(0 , "Employee_ID" , [101, 102, 103, 104, 105, 106])
print("\nDataFrame after inserting Employee_ID column:")
print(df)
```

```python
data = {
    "name":['Raj Sharma', 'Surya Pratap', 'Prince Kumar', 'Aniket Yadav', 'Ayush Singh', 'Prince sharma'],
    "age": [19, 20, 21, 32, 24, 22],
   "salray": [20000, 25000, 30000, 40000, 35000, 32000],
    "performance score": ['Good', 'Very Good', 'Excellent', 'Average', 'Good', 'Excellent']
}
df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)

 # loc[] - > to add column at specific location
 # df.loc[<row_index> , "<column_name>"] = <some_value>
df.loc[0 , "salray"] = 22000
print("\nDataFrame after updating salray for first employee:")
print(df)
```

Removing missing data :

```python
data = {
    "name":['Raj Sharma', None, 'Prince Kumar', 'Aniket Yadav', 'Ayush Singh', 'Prince sharma'],
    "age": [19, None, 21, 32, 24, 22],
   "salray": [20000, None, 30000, 40000, 35000, 32000],
    "performance score": ['Good', None, 'Excellent', 'Average', 'Good', 'Excellent']
}
df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)

df.dropna(inplace=True)
print("\nDataFrame after dropping rows with any null values:")
print(df)
```

Adding / filling missing data :

```python
data = {
    "name":['Raj Sharma', None, 'Prince Kumar', 'Aniket Yadav', 'Ayush Singh', 'Prince sharma'],
    "age": [19, None, 21, 32, 24, 22],
   "salray": [20000, None, 30000, 40000, 35000, 32000],
    "performance score": ['Good', "V.Good", 'Excellent', 'Average', 'Good', 'Excellent']
}
df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)

df['age'].fillna(df['age'].mean(), inplace=True)
df['salray'].fillna(df['salray'].mean(), inplace=True)
print("\nDataFrame after filling null values with mean:")
print(df)
```

interpolate()

```python
# interpolation to fill null values 
   # types of interpolation : linear, time, index, values, nearest,
                             #  zero, slinear, quadratic, cubic, spline, polynomial, barycentric, krog

data = { 
    "Time": [1, 2, 3, 4, 5, 6],
    "Value" : [10, None, 30, None, 50, 60]
}
df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)

df["Value"] = df["Value"].interpolate(method='linear')
print("\nDataFrame after linear interpolation:")
print(df)
```

Sorting 

```python
data= {
    "name":['Raj Sharma', 'Surya', 'Prince Kumar', 'Aniket Yadav', 'Ayush Singh', 'Prince sharma'],
    "age": [19, 8, 21, 32, 24, 22],
   "salray": [20000, 212, 30000, 40000, 35000, 32000],
}
df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)
df.sort_values(by="age",ascending=False , inplace=True)
print("\nDataFrame after sorting by age:")
print(df)

#mutli-level sorting
df.sort_values(by=["age", "salray"], inplace=True)
print("\nDataFrame after multi-level sorting by age and salray:")
print(df)
```

Grouping

```python
data= {
    "name":['Raj Sharma', 'Surya', 'Prince Kumar', 'Aniket Yadav', 'Ayush Singh', 'Prince sharma'],
    "age": [19, 19, 22, 32, 24, 22],
   "salary": [20000, 212, 30000, 40000, 35000, 32000],
}
df = pd.DataFrame(data)
print("Original DataFrame:")    
print(df)
grouped = df.groupby('age')["salary"].sum()
print("\nSum of salaries grouped by age:")
print(grouped)
```

Merging

```python
data_Customers = pd.DataFrame({
    "CustomerID": [1, 2, 3, 4, 5, 6],
    "Region": ['North', 'South', 'East', 'West', 'North', 'South'],
    "Sales": [250, 150, 400, 300, 200, 350]
})
dat_order = pd.DataFrame({
    "OrderID": [101, 102, 103, 104, 105, 106],
    "CustomerID": [1, 2, 3, 4, 15, 6],
    "OrderAmount": [500, 300, 700, 400, 600, 800]
})
mergeInnwe = pd.merge(data_Customers, dat_order, on="CustomerID", how="inner") # Matching rows only
print("\nInner Merged DataFrame:")
print(mergeInnwe)
mergeLeft = pd.merge(data_Customers, dat_order, on="CustomerID", how="left") # All left + matches
print("\nLeft Join Merged DataFrame:")
print(mergeLeft)
mergeRight = pd.merge(data_Customers, dat_order, on="CustomerID", how="right") # All right + matches
print("\nRight Join Merged DataFrame:")
print(mergeRight)
mergeOuter = pd.merge(data_Customers, dat_order, on="CustomerID", how="outer") # All rows from both
print("\nOuter Join Merged DataFrame:")
print(mergeOuter)

```

Concatination 

```python
Region1 = pd.DataFrame({
    "customer_Id": [1, 2, 3],
    "Name": ['Raj', 'Surya', 'Prince']
})
Region2 = pd.DataFrame({
    "customer_Id": [4, 5, 6],
    "Name": ['Aniket', 'Ayush', 'Aman']
})  
combined_vertical = pd.concat([Region1, Region2], ignore_index=True)
print("\nCombined DataFrame vertically using concat():")   
print(combined_vertical)
combined_horizontal = pd.concat([Region1, Region2], axis=1)
print("\nCombined DataFrame horizontally using concat():")
print(combined_horizontal)
```
